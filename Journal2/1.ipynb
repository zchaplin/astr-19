Create a Jupyter notebook, import matplotlib. Write
cells that create an array x ranging from [0,1] in 100 steps and that defines
a function that returns exp(x). In a new cell use the function to set
y=exp(x), and then plot x vs. y. Label the x-axis as “Time [milliseconds]”
and the y-axis as “Awesomeness”. Save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(1, 2, 100)
def create():
    y = np.exp(x)
    return y
def math():
    y = create()
    plt.plot(x, y)
    plt.xlabel("milliseconds")
    plt.ylabel("awesomeness")
    plt.title("Awesomeness vs. Time")
    plt.savefig("plot.pdf")  # Save the figure as a PDF
    plt.show()

math()
    
    
```
