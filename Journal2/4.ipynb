```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.exponential(scale=3.0, size=1000)
```

```python
plt.hist(random_numbers, bins=100, color='red', edgecolor='pink')

# Label the axes
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Exponetial")

# Save the figure as a PDF
plt.savefig("uniform_histogram.pdf")
plt.show()
```
