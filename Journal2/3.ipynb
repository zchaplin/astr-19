Create a Jupyter notebook, import matplotlib. Use
numpy to pull 1000 random numbers distributed uniformly between [0,1].
Histogram the random numbers into 100 bins, and plot the histogram.
Label your axes and save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.uniform(0, 1, 1000)
```

```python
plt.hist(random_numbers, bins=100)

# Label the axes
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Random Numbers")

# Save the figure as a PDF
plt.savefig("uniform_histogram.pdf")
plt.show()
```
